```python
import numpy as np
import pandas as pd
from sklearn.datasets import * 
import matplotlib.pyplot as plt

from sklearn.metrics import mean_absolute_error, mean_squared_error, median_absolute_error, r2_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv('../datasets/heart.csv')
```

```python
dataCopy = data.copy()
```

Кодирование категориальных признаков

```python
data['slope'] = data['slope'].astype('category')
data['ca'] = data['ca'].astype('category')
data['thal'] = data['thal'].astype('category')
```

Назначение закодированной переменной новому столбцу

```python
data['slope_category'] = data['slope'].cat.codes
data['ca_category'] = data['ca'].cat.codes
data['thal_category'] = data['thal'].cat.codes
```

```python
data = data.drop(['slope', 'ca', 'thal'], axis=True)
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
data=data.dropna()
data.isna().sum()
```

```python
data.describe().T
```

```python
y=data['sex']
x=data.drop('sex', axis=True)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=45)
```

```python
x_train.shape, y_train.shape
```

```python
x_test.shape, y_test.shape
```

```python
np.unique(y_train)
```

```python
np.unique(y_test)
```

Метрики построенной модели

```python
scaler=MinMaxScaler().fit(x_train)
x_train=pd.DataFrame(scaler.transform(x_train), columns=x_train.columns)
x_test=pd.DataFrame(scaler.transform(x_test), columns=x_test.columns)
x_train.describe()
```

```python
def test_model(model):
    print("mean_absolute_error: ", mean_absolute_error(y_test, model.predict(x_test)))
    print("mean_squared_error: ", mean_squared_error(y_test, model.predict(x_test)))
    print("median_absolute_error: ", median_absolute_error(y_test, model.predict(x_test)))
    print("r2_score: ", r2_score(y_test, model.predict(x_test)))
```

Метод ближайших соседей с гиперпараметром K=2

```python
reg_2=KNeighborsRegressor(n_neighbors=2)
reg_2.fit(x_train, y_train)
```

```python
test_model(reg_2)
```

mean_absolute_error: 0.066, чем ближе к нулю, тем выше качество регрессии

mean_squared_error:  0.043, чем ближе к нулю, тем лучше модель

median_absolute_error:  0.0
